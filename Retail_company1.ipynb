```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")

print(df.head())
```

```python
df["Total Sales"] = df["Quantity"] * df["Price per Unit"]
```

```python
print(df.head())
```

```python
total_revenue = df["Total Sales"].sum()
print("Total Revenue:", total_revenue)
```

```python
categories = df["Category"].unique()
print("Categories:", categories)
```

```python
region_sales = df.groupby("Region")["Total Sales"].sum()
top_region = region_sales.idxmax()

print("Top Region:", top_region)
```

```python
  avg_quantity = df.groupby("Category")["Quantity"].mean()
print(avg_quantity)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

daily_sales = df.groupby("Order Date")["Total Sales"].sum()
best_day = daily_sales.idxmax()

print("Best Day:", best_day)
```

```python
filtered = df[df["Quantity"] > 2]
print(filtered)
```

```python
region_sales.plot(kind="bar")
plt.title("Total Sales by Region")
plt.show()
```

```python
sorted_df = df.sort_values(by="Total Sales", ascending=False)
print(sorted_df.head())
```
